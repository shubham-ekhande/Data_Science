```python
## Used to test whether three or more group means are different.
```

```python
import numpy as np
import pandas as pd
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
## One-way ANOVA example
method1 = [85, 88, 82, 86, 89]
method2 = [78, 82, 80, 79, 81]
method3 = [92, 95, 90, 93, 94]

f_stat, p_value = stats.f_oneway(method1, method2, method3)
print(f"ANOVA: F = {f_stat:.3f}, p = {p_value:.3f}")
```

```python
## Post-hoc test (Tukey HSD) if ANOVA significant
data = method1 + method2 + method3
groups = ['Method1']*5 + ['Method2']*5 + ['Method3']*5
tukey_result = pairwise_tukeyhsd(data, groups,
alpha=0.05)
print("\nTukey HSD Post-hoc Test:")
print(tukey_result)
```
