```python
#Sorting
df.sort_values('TotalCharges',ascending=False)
```

```python
## Applying functions
df['ChurnFlag'] = df['Churn'].map({'Yes': 1, 'No': 0})
df['Charges_Category'] = df['MonthlyCharges'].apply(lambda x: 'High' if x >
80 else 'Low')
```

```python
# Groupby and aggregation
df.groupby('gender')['MonthlyCharges'].mean()
df.groupby('Contract')['ChurnFlag'].agg(['mean', 'count'])
```

```python
# Pivot table
pd.pivot_table(df, values='MonthlyCharges', index='InternetService',
columns='Churn', aggfunc='mean')
```

```python
# Concatenation example
df1 = df.head(10)
df2 = df.tail(10)
df_concat = pd.concat([df1, df2])
```

```python
# Adding new column
df['TotalRevenue'] = df['MonthlyCharges'] * df['tenure']
```

```python
# Merge example (dummy example)
customer_info = df[['gender', 'SeniorCitizen']]
billing = df[['MonthlyCharges', 'TotalRevenue']]
merged_df = pd.concat([customer_info, billing], axis=1)
```

```python
#Working with dates
df['InvoiceDate'] = pd.date_range(start='2022-01-01', periods=len(df),
freq='D')
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
df['Day'] = df['InvoiceDate'].dt.day
```

```python
#Advanced ops
# Crosstab
pd.crosstab(df['gender'], df['Churn'])
# Rolling mean (Window function)
df['RollingAvgCharges'] = df['MonthlyCharges'].rolling(window=5).mean()
```
