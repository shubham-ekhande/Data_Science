```python
## Used to compare sample mean with known population mean when:
    # Population standard deviation (σ) is known, and
    # Sample size n ≥ 30.
```

```python
import numpy as np
from scipy import stats
```

```python
def z_test(sample, pop_mean, pop_std):
    sample_mean = np.mean(sample)
    n = len(sample)

    z_value = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p_value = 2 * (1- stats.norm.cdf(abs(z_value)))

    print(f"Z-test: Z = {z_value:.3f}, p = {p_value:.3f}")

sample_data = [102, 105, 98, 107, 103, 101, 99, 104, 106, 100]
z_test(sample_data, 100, 15)
```
