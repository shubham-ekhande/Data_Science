```python
##Section C Long Quetions
#Q6. Describe the difference between Scatter plot, Line plot, and Histogram in
#terms of:
# Purpose
# When to use
# Example from the dataset
# Visualization insights

1. Scatter Plot:
    Purpose: To see the relationship between two things.
    When to use: When you want to check if one variable changes as another changes, or to find patterns, outliers.
    Example from dataset: Plot likes vs comments to see if posts with more likes also get more comments.
    Insight: Shows trends or unusual points. For example, a post with many likes but very few comments may stand out.

2. Line Plot:
    Purpose: To show how something changes over time.
    When to use: When you want to track trends day by day, month by month, or over any continuous period.
    Example from dataset: Plot post_day or post_time vs engagement to see how engagement changes daily.
    Insight: Helps spot trends and patterns. For example, engagement might rise on weekends or drop during certain periods.

3. Histogram:
    Purpose: To see how values are distributed.
    When to use: When you want to know which values are common, rare, or if the data is skewed.
    Example from dataset: Plot likes frequency to see how many posts fall into low, medium, or high likes.
    Insight: Shows data distribution. For example, most posts may get low likes, and only a few get very high likes.
```

```python
#Q7. Explain how Matplotlib styling (plt.style.use) transforms readability.
#Include:
# Why styles matter
# Example using dark mode/light mode
# How style improves dashboards and client presentations

1. Why Styles Matter:
    Styling changes the look and feel of your plots—colors, grids, fonts, and backgrounds.
    A well-styled plot is easier to read, highlights important data, and avoids clutter.
    Without styling, plots may look plain, confusing, or hard to interpret, especially 
    for presentations or dashboards.

2. Example: Dark Mode vs Light Mode
    Light mode: White background, dark lines, classic look. Good for reports or printed slides.

        import matplotlib.pyplot as plt
        plt.style.use('seaborn-whitegrid')  # Light mode
        plt.plot([1,2,3,4], [10,20,15,25])
        plt.title("Light Mode Example")
        plt.show()


    Dark mode: Dark background, bright lines, modern look. Good for dashboards or screen presentations.

        plt.style.use('dark_background')  # Dark mode
        plt.plot([1,2,3,4], [10,20,15,25], color='cyan')
        plt.title("Dark Mode Example")
        plt.show()

3. How Styles Improve Dashboards and Client Presentations:
    Makes plots visually appealing and professional.
    Highlights key trends with clear contrasts.
    Consistent styling across charts helps clients understand insights quickly.
    Example: Using a dark mode for a live engagement dashboard makes peaks and drops stand out immediately.
```
