```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("social_media_engagement1.csv")
df.head()
```

```python
#A marketing team wants to understand how likes are distributed across all posts.
#Write code to:
# Load dataset
# Create sns.histplot() with KDE enabled
# Add title, labels, color
# Write insight on distribution shape

sns.histplot(x='likes',kde=True,data=df,bins=30,color='magenta')
plt.title('Distribution of Likes Across Posts')

plt.show()
```

```python
#The brand wants to compare average reach based on content type.
#Write Seaborn code to:
# Group by Content_Type
# Use sns.barplot()
# Add palette
# Display which content type performs best

df['reach'] = df['likes'] + df['comments'] + df['shares']
avg_reach = df.groupby('post_type')['reach'].mean()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=avg_reach.index,y=avg_reach.values)
plt.title('Average Reach by Content Type', fontsize=16)
plt.xlabel('Content Type', fontsize=12)
plt.ylabel('Average Reach', fontsize=12)

plt.show()
```

```python
#Find how content length influences engagement.
#Tasks:
# Create sns.regplot() for Content_Length vs Engagement
# Add color and title
# Interpret if relationship is positive/negative

# Assuming engagement is sum of likes, comments, shares
df['engagement'] = df['likes'] + df['comments'] + df['shares']

# Plot regression between Content_Length and Engagement
plt.figure(figsize=(10,6))
sns.regplot(x='post_type', y='engagement', data=df, color='green')

# Add title and labels
plt.title('Relationship Between Content Length and Engagement', fontsize=16)
plt.xlabel('Content Length', fontsize=12)
plt.ylabel('Engagement', fontsize=12)

plt.show()
```

```python
#Management wants a multi-relationship overview.
#Write code to:
# Create sns.pairplot() on Likes, Comments, Shares
# Add hue="Content_Type"
# Explain any two patterns seen
 # vars cause we only want to take like,share,comments
sns.pairplot(df,vars=['likes','comments','shares'],hue='post_type')
plt.suptitle('Pairwise Relationships Between Likes, Comments, and Shares', fontsize=16, y=1.02)
plt.show()
```

```python
#The analyst wants to check which numeric variable correlates most withengagement.
#Tasks:
# Compute correlation matrix
# Plot using sns.heatmap(annot=True, cmap="coolwarm")
# Identify and discuss two strong correlations

corr = df.corr(numeric_only=True)

sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("Heatmap - Correkatin Matrix")

plt.show()
```

```python
## Section C Long Answers
#Q6. Explain difference between: (with examples)
# barplot vs countplot
# histplot vs kdeplot
# regplot vs scatterplot
#Write purpose + example from your dataset.

Barplot vs Countplot:
    Barplot: Shows numeric values per category.
    Example: Average likes per post type  sns.barplot(x='post_type', y='likes', data=df)
    Countplot: Shows number of items in each category.
    Example: Number of posts per type  sns.countplot(x='post_type', data=df)

Histplot vs KDEplot:
    Histplot: Shows how numeric values are distributed using bins.
    Example: Distribution of likes  sns.histplot(df['likes'], bins=20)
    KDEplot: Shows smoothed distribution of data.
    Example: Likelihood of likes falling in a range  sns.kdeplot(df['likes'], shade=True)

Scatterplot vs Regplot:
    Scatterplot: Plots points for two numeric variables.
    Example: Likes vs content length  sns.scatterplot(x='content_length', y='likes', data=df)
    Regplot: Scatterplot with trend line.
    Example: Likes vs content length with regression  sns.regplot(x='content_length', y='likes', data=df)
```

```python
#Q7. Why Seaborn is ideal for statistical storytelling?
#Explain with:
# Real example
# Visual clarity
# Built-in themes
# Automatic statistical summaries
# Comparison with Matplotlib

Real example:
    Seaborn makes it easy to explore social media data. 
    For example, sns.barplot(x='post_type', y='likes', data=df) 
    can instantly show which post type gets the most engagement.


Visual clarity:
    Plots are clean, colorful, and easy to read without extra formatting. 
    For instance, a sns.boxplot automatically shows median, quartiles, and outliers clearly.
    

Built-in themes:
    Seaborn has themes like "darkgrid", "whitegrid", "ticks", which improve aesthetics quickly:
        sns.set_style("whitegrid")
        sns.histplot(df['likes'])


Automatic statistical summaries:
    Functions like barplot or pointplot compute and show summary statistics (mean, confidence intervals) automatically, 
    saving manual calculations.


Comparison with Matplotlib:
    Seaborn: Focused on statistics and visual storytelling; fewer lines of code; automatically adds clarity and summaries.
    Matplotlib: More flexible for custom plotting, but requires more coding for styling and statistical details
```
