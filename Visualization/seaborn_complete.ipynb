```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# Load Seaborn built-in dataset
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
# 1 Line plot
sns.lineplot(x="total_bill", y="tip", data=df, color="blue", marker="o")
plt.title("Line Plot - Total Bill vs Tip")
plt.show()
```

```python
# 2 Bar Plot
sns.barplot(x="day", y="total_bill", data=df, palette="pastel",hue="day")
# or we can use this : sns.barplot(x="day",y="total_bill",hue="sex",data=df,palette="pastel")
plt.title("Bar Plot - Average Bill by Day")
plt.show()
```

```python
# 3 Count Plot
sns.countplot(x="day", data=df, palette="Set2",hue='sex')
plt.title("Count Plot - Number of Records per Day")
plt.show()
```

```python
# 4 Histogram
sns.histplot(df["total_bill"], bins=20, kde=True, color="coral")
plt.title("Histogram - Distribution of Total Bill")
plt.show()
```

```python
# 5 KDE plot
sns.kdeplot(df["total_bill"], fill=True, color="green")
plt.title("KDE Plot - Density of Total Bill")
plt.show()
```

```python
# 6 Box Plot
sns.boxplot(x="day", y="total_bill", data=df, palette="Set3")
plt.title("Box Plot - Spread of Total Bill by Day")
plt.show()
```

```python
# 7 Violin Plot
sns.violinplot(x="day", y="total_bill", data=df, palette="muted")
plt.title("Violin Plot - Distribution by Day")
plt.show()
```

```python
# 8 Strip Plot
sns.stripplot(x="day", y="total_bill", data=df, jitter=True, color="red")
plt.title("Strip Plot - Individual Data Points")
plt.show()
```

```python
# 9 Swarm Plot
sns.swarmplot(x="day", y="total_bill", data=df, palette="cool")
plt.title("Swarm Plot - Non-overlapping Data Points")
plt.show()
```

```python
# 10 Pair Plot
sns.pairplot(df, hue="sex", palette="husl")
plt.suptitle("Pair Plot - Relationships Between Features", y=1.02)
plt.show()
```

```python
# 11 HeatMap
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Heatmap - Correlation Matrix")
plt.show()
```

```python
# 12 Joint Plot
sns.jointplot(x="total_bill", y="tip", data=df, kind="reg", color="purple")
plt.suptitle("Joint Plot - Total Bill vs Tip", y=1.02)
plt.show()
```

```python
# 13 Cat Plot
sns.catplot(x="day", y="total_bill", kind="bar", data=df, palette="dark")
plt.title("Cat Plot - Bar Kind")
plt.show()

sns.catplot(x="day", y="total_bill", kind="box", data=df, palette="deep")
plt.title("Cat Plot - Box Kind")
plt.show()
```

```python
# 14 Relational Plots
sns.relplot(x="total_bill", y="tip", hue="sex", size="size", data=df,palette="dark")
plt.title("Relational Plot - Total Bill vs Tip")
plt.show()
```

```python
# 15 Regression Plot
sns.regplot(x="total_bill", y="tip", data=df, color="teal")
plt.title("Regression Plot - Linear Relationship")
plt.show()
```

```python
# 16 Distribution Plot
sns.displot(df["total_bill"], kde=True, color="gold")
plt.title("Distribution Plot - Total Bill")
plt.show()
```

```python
# 17 Theme / Style Examples
sns.set_style("darkgrid")
sns.boxplot(x="day", y="total_bill", data=df, palette="Set2")
plt.title("Box Plot - Darkgrid Style")
plt.show()
sns.set_style("white")
sns.violinplot(x="day", y="tip", data=df, palette="husl")
plt.title("Violin Plot - White Style")
plt.show()
```
