```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
df = pd.read_csv("house.csv")
```

```python
df.head()
```

```python
df = df.dropna()
```

```python
## Model Building
x = df[['total_rooms']]
y = df['median_house_value']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(x_train,y_train)

y_pred = model.predict(x_test)

print("Slope (b1):", model.coef_)
print("Intercept (b0):", model.intercept_)
print("R²:", r2_score(y_test, y_pred))
```

```python
## prediction
pred = pd.DataFrame([{'total_rooms' : 250}])
print("Price of house with 250 Rooms:",int(model.predict(pred)[0]))
```

```python
## Multi_linear_regression
X_multi = df[['housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income']]
#X = df[features]
y_multi = df['median_house_value']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_multi, y_multi, test_size = 0.2)

multi_model = LinearRegression()
multi_model.fit(X_train,y_train)

y_pred_m = multi_model.predict(X_test)

print(r2_score(y_test, y_pred_m))
```

```python
features = pd.DataFrame([{'housing_median_age' : 30 , 
              'total_rooms' : 1500, 
              'total_bedrooms' : 500, 
              'population' : 2000, 
              'households' : 700, 
              'median_income' : 8.3225
             }])

print("Predicting House Price: ",int(multi_model.predict(features)[0]))
```
