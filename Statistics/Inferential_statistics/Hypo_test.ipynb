```python
from scipy import stats
import numpy as np
# One-sample t-test example
# Claim: Average student IQ = 100
sample_iq = [105, 98, 112, 107, 95, 103, 110, 102, 108,115]

population_mean = 100
t_stat, p_value = stats.ttest_1samp(sample_iq, population_mean)
alpha = 0.05
print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.3f}")
if p_value < alpha:
    print("Reject H₀: Average IQ is significantly different from 100")
else:
    print("Fail to reject H₀: No evidence that average IQ differs from 100")
```
