```python
# To test whether two populations have equal variances. Often used before a t-test or in ANOVA.
# Varience Comparison
```

```python
import numpy as np
from scipy import stats
```

```python
def f_test(group_1, group_2):
    var_1 = np.var(group_1, ddof=1)
    var_2 = np.var(group_2, ddof=1)

    #Ensure larger varience in numerator
    if var_1 > var_2:
        f_value = var_1 / var_2
        df1 = len(group_1) - 1
        df2 = len(group_2) - 1
    else:
        f_value = var_2 / var_1
        df1 = len(group_2) - 1
        df2 = len(group_1) - 1

    p_value = 1 - stats.f.cdf(f_value, df1, df2)
    print(f"f_value: {f_value:.3f}, p_value: {p_value:.3f}")

group_a = [85, 88, 92, 78, 83, 90, 87, 84, 89, 86]
group_b = [78, 82, 85, 79, 80, 83, 81, 77, 84, 79]

f_test(group_a, group_b)
        
```
