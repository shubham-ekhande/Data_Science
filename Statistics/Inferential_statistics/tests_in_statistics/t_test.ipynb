```python
import numpy as np
from scipy import stats
```

```python
## Independent t_test
# Used to compare two separate groups
group_1 = [85, 88, 92, 78, 83, 90, 87, 84, 89, 86]
group_2 = [78, 82, 85, 79, 80, 83, 81, 77, 84, 79]

t_stat, p_value = stats.ttest_ind(group_1, group_2)
print(f"t_value: {t_stat:.3f}, p_value: {p_value:.3f}")
```

```python
## Paired t_test (before vs after)
# Used when same group is measured before and after some intervention (e.g.,training, treatment).
before = [70, 72, 68, 75, 71, 69, 74, 73, 70, 72]
after = [75, 78, 72, 80, 76, 74, 79, 77, 75, 78]

t_stat,p_value = stats.ttest_rel(before, after)
print(f"t_value: {t_stat:.3f}, p_value: {p_value:.3f}")
```
