```python
# 3. Non-Probability Sampling (Non-random)
# Here, not everyone has a chance. Selection depends on convenience or judgment
# like manually sampling 
```

```python
# (a) Convenience Sampling
#Choose who’s easy to reach.
sample = ['he','she','it']  # easily available
sample
```

```python
# (b) Purposeful Sampling
# Choose specific people with particular characteristics.

data = {
    'name' : ['a','b','c'],
    'marks' : [95, 40, 88]
}
# selecting those with marks higher than 80

high = [d for d in data['name'] if data['marks'][data['name'].index(d)] > 80]
high
```

```python
# (c) Snowball Sampling
# Start with one person → they refer others.
chain = ['p1']
chain += ['p2','p3']      # its like reffered
chain
```
