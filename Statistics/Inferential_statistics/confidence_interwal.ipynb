```python
import numpy as np
import scipy.stats as stats

# NumPy: Used for numerical operations like mean and standard deviation.
# SciPy.stats: Provides statistical tools like t-distribution functions.
```

```python
heights = [168, 172, 175, 169, 171, 174, 170, 173, 167, 172]

confidence_level = 0.95
```

```python
## calulate CI
# sample mean
sample_mean = np.mean(heights)
sample_mean
```

```python
# sample standard_deviation
sample_std = np.std(heights, ddof=1)  # ddof=1 for sample standard deviation
sample_std
```

```python
n = len(heights)
```

```python
# Critical Value
t_value = stats.t.ppf((1+ confidence_level) / 2, df=n-1)
round(t_value,5)
```

```python
# Margin of error
margin_of_error = t_value * (sample_std / np.sqrt(n))
margin_of_error
```

```python
# Computing cnfidence intervals
ci_upper = sample_mean + margin_of_error
ci_lower = sample_mean - margin_of_error
print(f"95% CI: ({ci_lower:.2f}, {ci_upper:.2f})")
```
