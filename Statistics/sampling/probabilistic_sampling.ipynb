```python
import pandas as pd
import random
```

```python
# 1. Population and Sampling
"""
    Theory
        Population → the entire group you want to study.
            Example: All students in your college.
        Sample → a small subset of the population used for analysis.
            Example: 50 students chosen from the college.
            We sample because testing everyone is expensive/time-consuming.
"""
```

```python
# 2. Probability Sampling (Random methods)
"""These are random, meaning every member has some known equal chance of being selected."""
```

```python
# (a) Simple Random Sampling
#(a) Simple Random Sampling
#Each person/item has an equal chance.
#Example:
#Pick 5 students randomly from 100.

population = list(range(1,101))
sample = random.sample(population,5)        #Everytime random samples are generated
sample
```

```python
# (b) Systematic Sampling
#Pick every k-th item.
#Example:
#If you want 10 samples out of 100 → pick every 10th.

population = list(range(1,101))
k = 10
sample = population[::k]   # Here every 10th elemet is picked (k=10)
sample
```

```python
# (c) Stratified Sampling
#Divide the population into groups (strata), then randomly sample from each group.
#Example:
#Divide students by branch (CSE, ECE, MECH) and sample from each.

students = {
    'cse' : ['a1','a2','a3','a4'],
    'ece' : ['b1','b2','b3','b4'],
    'mech' : ['c1','c2','c3','c4']
}

sample = []
for branch, members in students.items():
    sample += random.sample(members, 2)  # 2 items from each class
sample
```

```python
# (d) Cluster Sampling
# Randomly select entire groups (clusters).
#Example:
#Choose 2 classes (clusters) out of 10 and survey all students in those classes. 

clusters = {
    'class_1' : ['s1','s2','s3'],
    'class_2' : ['s4','s5','s6'],
    'class_3' : ['s7','s8','s9']
}

selected_clusters = random.sample(list(clusters.keys()),1)  #here one cluster is selected for sampling
print(selected_clusters)
sample = []
for c in selected_clusters:
    for student in clusters[c]:
        sample.append(student)
sample

#or
#sample = [student for c in selected_clusters for student in clusters[c]]  #list comprehension
```
