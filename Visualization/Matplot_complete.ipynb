```python
# Matplotlib â€” All Major Visualizations
import matplotlib.pyplot as plt
import numpy as np
# Sample data
x = np.arange(1, 6)
y = np.array([10, 15, 13, 18, 16])
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 78, 33]
```

```python
x
```

```python
# 1 LINE PLOT
plt.figure()
plt.plot(x, y, color='purple', marker='o', linestyle='--')
plt.title("Line Plot - Trend Over Time")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
# 2 BAR CHART
plt.figure()
plt.bar(categories, values, color='skyblue')
plt.title("Bar Chart - Category Comparison")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

```python
# 3 HORIZONTAL BAR CHART

plt.figure()
plt.barh(categories, values, color='orange')
plt.title("Horizontal Bar Chart")
plt.xlabel("Value")
plt.ylabel("Category")
plt.show()
```

```python
# 4 PIE CHART

plt.figure()
plt.pie(values, labels=categories, autopct='%1.1f%%',
startangle=90)
plt.title("Pie Chart - Category Share")
plt.show()
```

```python
# 5 SCATTER PLOT

plt.figure()
x_scatter = np.random.randint(1, 100, 50)
y_scatter = np.random.randint(1, 100, 50)
plt.scatter(x_scatter, y_scatter, color='green',marker='*')
plt.title("Scatter Plot - Relationship Between Variables")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# 6 HISTOGRAM
plt.figure()
data = np.random.randn(1000)
plt.hist(data, bins=20, color='teal',
edgecolor='black')
plt.title("Histogram - Data Distribution")
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.show()
```

```python
# 7 BOX PLOT
plt.figure()
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 15, 100)
plt.boxplot([data1, data2], patch_artist=True,
boxprops=dict(facecolor='lightgreen'))
plt.title("Box Plot - Data Spread and Outliers")
plt.xlabel("Dataset")
plt.ylabel("Values")
plt.show()
```

```python
data1
```

```python
# 8 STACK PLOT

plt.figure()
days = [1, 2, 3, 4, 5]
a = [3, 4, 5, 4, 3]
b = [1, 2, 3, 3, 2]
c = [2, 3, 4, 4, 3]
plt.stackplot(days, a, b, c, labels=['A', 'B', 'C'],
colors=['r', 'g', 'b'])
plt.title("Stack Plot - Cumulative Data Over Time")
plt.xlabel("Day")
plt.ylabel("Value")
plt.legend()
plt.show()

```

```python
# 9 SUBPLOTS (Multiple Graphs Together
fig, axs = plt.subplots(2, 2, figsize=(8, 6))
axs[0, 0].plot(x, y, color='blue')
axs[0, 0].set_title("Line Plot")
axs[0, 1].bar(categories, values, color='green')
axs[0, 1].set_title("Bar Chart")
axs[1, 0].hist(data, bins=10, color='red')
axs[1, 0].set_title("Histogram")
axs[1, 1].scatter(x, y, color='purple')
axs[1, 1].set_title("Scatter Plot")
plt.tight_layout()
plt.show()
```

```python
# 10 AREA PLOT

plt.figure()
x_area = np.linspace(0, 10, 100)
y_area = np.sin(x_area)
plt.fill_between(x_area, y_area, color='skyblue',
alpha=1)
plt.title("Area Plot - Filled Line Area")
plt.xlabel("X-axis")
plt.ylabel("sin(x)")
plt.show()
```

```python
# 11 STEM PLOT

plt.figure()
x_stem = np.arange(1, 10)
y_stem = np.random.randint(1, 20, len(x_stem))
plt.stem(x_stem, y_stem, linefmt='grey', markerfmt='o',
basefmt='k')
plt.title("Stem Plot - Discrete Data")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# 12 ERROR BAR PLOT

plt.figure()
x_err = np.arange(1, 6)
y_err = np.random.randint(10, 20, 5)
error = np.random.rand(5)
plt.errorbar(x_err, y_err, yerr=error, fmt='o',
color='brown', capsize=5)
plt.title("Error Bar Plot - Uncertainty in Data")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# 13 ANNOTATION EXAMPLE
plt.figure()
plt.plot(x, y, marker='o', color='darkblue')
plt.title("Annotation Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.annotate("Highest Point", xy=(x[3], y[3]),
xytext=(x[3]+0.2, y[3]+1),
arrowprops=dict(facecolor='black',
arrowstyle='->'))
plt.show()
```

```python
# 14 CUSTOM STYLE (Themes)

plt.style.use('ggplot')
plt.figure()
plt.plot(x, y, color='crimson', marker='o')
plt.title("Styled Line Plot - ggplot Theme")
plt.show()
# Reset style
plt.style.use('default')
```
