```python
## Used for categorical data to test:
# Independence between two variables, or
# Goodness of fit of observed vs expected frequencies.
```

```python
import numpy as np
from scipy.stats import chi2_contingency
```

```python
# Example: Gender vs Voting Preference
observed = [
    [40, 60], # Male: Party A, Party B
    [70, 30] # Female: Party A, Party B
]

chi2, p_value, dof, expected = chi2_contingency(observed)
print(f"Chi-squre test: χ² : {chi2:.3f}, p : {p_value:.3f}")
print("Expected frequencies:")
print(expected)
```
