```python
# 3. Indexing,Selection,Filtering
""" Pandas provides powerful ways to select rows and columns.
 Main methods:
1. [] → Select column(s) or row slices
2. .loc[] → Label-based selection
3. .iloc[] → Position-based selection
4. Boolean/conditional filtering"""
```

```python
# A) Selecting columns
import pandas as pd



df = pd.DataFrame(data)

#selcting single column
print(df["name"]) #or print(df[["name"]])
print("\n")
#selecting multiple rows
print(df[["name","age","marks"]])
```

```python
# B) selecting rows with slicing
import pandas as pd


print(df[1:6])
```

```python
# C) Using .loc[] (label-based)

import pandas as pd

#single row by index label
print(df.loc[1])
print("\n")

#multiple rows
print(df.loc[[1,3]])
print("\n")

#Rows and column selection
print(df.loc[0:2, ["name","marks"]])
```

```python
# D) Using iloc[] (position-based0)
print(df)
print("\n")

print(df.iloc[1])
print("\n")


print(df.iloc[1:4, 1:4])
```

```python
# E) Boolean Indexing

#student with marks > 80
print(df[df["marks"] > 80])
print("\n")

#students with age<25
print(df[df["age"] <= 23])
```

```python
# F) isin() and query()

#isin()
print(df[df["name"].isin(["Harsha","Shubham"])])
print("\n")

#query()
print(df.query("marks >= 90"))
```

```python
# 6. Data Cleaning in pandas
"""
    Raw data is often messy (missing values, duplicates, wrong column names, mixed data types). 
    Pandas provides tools to clean and prepare the data before analysis.
    Key operations:
        1. Handling missing values
        2. Handling duplicates
        3. Renaming columns
        4. Changing data types

"""
```

```python
# A). Handling Missing Values

#check missing values
print(df.isnull())
print("\n")

#check misssing values per column
print(df.isnull().sum())
print("\n")

#drops rows with missing values
print(df.dropna())
print("\n")

#Fill missing values with Zeros
print(df.fillna(0))
print("\n")

#fill with column mean (Average of column)
print(df["marks"].fillna(df["marks"].mean(), inplace=True))
print(df)
```

```python
# B). Handling Duplicates
import numpy as np
data = {
    "name":["Harshad","Harshad","Shubham","Rahul","Jay","Rohan","Soham","Tejas","Saket"],
    "age":[23,23,22,34,21,22,43,22,20],
    "marks":[43,43,56,76,89,99,76,87,33]
}
df = pd.DataFrame(data)
#check duplicates
print(df.duplicated())
print("\n")

#remove duplicates
print(df.drop_duplicates())
```

```python
# C). Renaming Columns
df.rename(columns={"name": "student_name", "marks": "score"}, inplace=True)
print(df)
```

```python
# D) Changing Data Types
# Convert Age from float to int
df["score"] = df["score"].astype("float")
print(df.dtypes)
```

```python
# E). Replacing Values

#Replace Single Value
df["stud_name"] = df["stud_name"].replace("Soham","Amit")
print(df)
print("\n")

#Replace Multiple Values
df["score"] = df["score"].replace({60,70,80,90,58,78,67,65})
print(df)
```
